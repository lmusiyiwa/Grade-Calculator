```python
# grade_calculator.ipynb

print("üéì Final Year ‚Äì 4th Semester Marks Evaluation üéì")
print("Bachelor of Psychology (Honours)\n")

# Inspirational education quote
print('"Education is the most powerful weapon which you can use to change the world." ‚Äì Nelson Mandela\n')

# Step 1: Get user input
name = input("Enter your full name: ").title().strip()

print(f"\nHello, {name}! Please enter your final semester marks below.")
print("These marks will determine whether you qualify to graduate in January next year.\n")

# Step 2: Get marks for each subject
behavioural_stats = float(input("Behavioural Statistics: "))
research_psych = float(input("Research Psychology: "))
psychometry = float(input("Psychometry: "))
integrated_practicum = float(input("Integrated Practicum: "))

# Store marks in a list
marks = [behavioural_stats, research_psych, psychometry, integrated_practicum]
subjects = ["Behavioural Statistics", "Research Psychology", "Psychometry", "Integrated Practicum"]

# Step 3: Calculate totals and average
total = sum(marks)
average = total / len(marks)

# Step 4: Determine grade
if average >= 80:
    grade = "A"
    comment = "Outstanding performance ‚Äî Dean‚Äôs List level!"
elif average >= 70:
    grade = "B"
    comment = "Excellent work ‚Äî a strong academic finish!"
elif average >= 60:
    grade = "C"
    comment = "Good effort ‚Äî consistent achievement!"
elif average >= 50:
    grade = "D"
    comment = "Satisfactory ‚Äî you‚Äôve met the minimum graduation criteria."
else:
    grade = "F"
    comment = "Unfortunately, this average is below the required level."

# Step 5: Determine pass/fail
result = "PASSED ‚úÖ" if average >= 50 else "FAILED ‚ùå"

# Step 6: Individual subject results
subject_results = ["Pass" if mark >= 50 else "Fail" for mark in marks]

# Step 7: Highest and lowest marks
highest = max(marks)
lowest = min(marks)

# Step 8: Display the final report
print("\nüìò --- Final Year Academic Report --- üìò")
print(f"Student: {name}")
for i in range(len(subjects)):
    print(f"{subjects[i]}: {marks[i]} ({subject_results[i]})")

print(f"\nTotal Marks: {total}")
print(f"Average: {average:.2f}")
print(f"Final Grade: {grade}")
print(f"Result: {result}")
print(f"Highest Mark: {highest}")
print(f"Lowest Mark: {lowest}")
print(f"\nFeedback: {comment}")

if average >= 75:
    print("‚ú® You qualify for Cum Laude honours ‚Äî an exceptional achievement!")
elif average >= 60:
    print("üéì You have performed well and are ready for professional practice.")
elif average >= 50:
    print("‚úÖ You‚Äôve successfully met graduation requirements.")
else:
    print("‚ö†Ô∏è Please consult your academic advisor for guidance.")

# Step 9: Graduation message
if result == "PASSED ‚úÖ":
    print("\nüéâ Congratulations! You are on track to graduate in January next year!")
else:
    print("\n‚ö†Ô∏è You may need to retake one or more modules before graduation.")

print("\nüìö 'Education is the passport to the future, for tomorrow belongs to those who prepare for it today.' ‚Äì Malcolm X")
```
